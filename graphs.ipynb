```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

```python
#Generating the tables from the text, skipping the header
stoke_unpol = np.genfromtxt('day2data4.txt',skip_header=19)
stoke_vert = np.genfromtxt('day2data5.txt',skip_header=19)
stoke_hori = np.genfromtxt('day2data6.txt',skip_header=19)
anti_stoke_hori = np.genfromtxt('day2data7.txt',skip_header=19)
anti_stoke_vert = np.genfromtxt('day2data8.txt',skip_header=19)
anti_stoke_unpol = np.genfromtxt('day2data9.txt',skip_header=19)
```

```python
#Creating a function to separate out the columns (which are the values we want)
def separate3col(array):
    col1 = []
    col2 = []
    col3 = []
    for i in range(0,len(array)):
        col1.append(array[i][0])
        col2.append(array[i][1])
        col3.append(array[i][2])
    return col1,col2,col3
```

```python
#Separating out the tables
stun_lambda,stun_int,stun_wnum = separate3col(stoke_unpol)
stvt_lambda,stvt_int,stvt_wnum = separate3col(stoke_vert)
sthr_lambda,sthr_int,sthr_wnum = separate3col(stoke_hori)
asthr_lambda,asthr_int,asthr_wnum = separate3col(anti_stoke_hori)
astvt_lambda,astvt_int,astvt_wnum = separate3col(anti_stoke_vert)
astun_lambda,astun_int,astun_wnum = separate3col(anti_stoke_unpol)

#I need to remove the last line from the anti-stokes horizontal polarized data
asthr_lambda.remove(asthr_lambda[-1])
asthr_int.remove(asthr_int[-1])
asthr_wnum.remove(asthr_wnum[-1])
```

```python
plt.scatter(stun_lambda,stun_int,s=1.75)
plt.plot(stun_lambda,stun_int,linestyle='-',linewidth=.5,c='r')
plt.title('Intensity vs Wavelength for Unpolarized Stokes Shift Emission')
plt.ylabel('Intensity')
plt.xlabel(r'Wavelength [$\AA$]')
```

```python
plt.scatter(stvt_lambda,stvt_int,s=1.75)
plt.plot(stvt_lambda,stvt_int,linestyle='-',linewidth=.5,c='r')
plt.title(r'Intensity vs Wavelength for Stokes Shift Emission at Polarization $\theta = 90^o$')
plt.ylabel('Intensity')
plt.xlabel(r'Wavelength [$\AA$]')
```

```python
plt.scatter(sthr_lambda,sthr_int,s=1.75)
plt.plot(sthr_lambda,sthr_int,linestyle='-',linewidth=.5,c='r')
plt.title(r'Intensity vs Wavelength for Stokes Shift Emission at Polarization $\theta = 0^o$')
plt.ylabel('Intensity')
plt.xlabel(r'Wavelength [$\AA$]')
print(sthr_lambda[sthr_int.index(max(sthr_int))])
```

```python
#Creating Relative Intensities
def rel_int(line):
    relative = []
    maximum = max(line)
    for i in line:
        daisy = round((i/maximum)*100, 1)
        relative.append(daisy)
    return relative
rel_stokeunpol = rel_int(stun_int)
rel_stokevert = rel_int(stvt_int)
rel_stokehori = rel_int(sthr_int)
rel_astokeunpol = rel_int(astun_int)
rel_astokevert = rel_int(astvt_int)
rel_astokehori = rel_int(asthr_int)
print(rel_stokeunpol.index(max(rel_stokeunpol)))
print(rel_stokeunpol[86])
print(rel_stokehori[86])
print(rel_stokevert[86])
```
